```python
import requests
from bs4 import BeautifulSoup
import csv


# Define the URL to get the links of individual car listings
#url = 'https://www.donedeal.ie/cars'
#response = requests.get(url)
#soup = BeautifulSoup(response.content, "html.parser")

# Find all the individual listings on the page
#listings = soup.find_all("li", {"class": "Listings__Desktop-sc-1igquny-3"})

# Prepare CSV file
with open("car_listing.csv", "w", newline="", encoding="utf-8") as csvfile:
    fieldnames = ["Price", "Year", "Make", "Model", "Mileage", "Transmission", "Fuel Type", "Body Type", "Color", "Engine Size", "NCT Expiry", "Road Tax"]
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()


    for i in range(800):  # Replace 10 with the desired number of loops
        start_num = i * 30+45000
        url = f"https://www.donedeal.ie/cars?start={start_num}"
        response = requests.get(url)
        soup = BeautifulSoup(response.content, "html.parser")
        listings = soup.find_all("li", {"class": "Listings__Desktop-sc-1igquny-3"})
        for item in listings:
            link = item.find("a", {"class": "Link__SLinkButton-sc-9jmsfg-0"})
            if link is not None and "https://www.donedeal.ie/cars-for-sale/" in link.get("href"):
                listing_url = link.get("href")
                print(f"Visiting {listing_url}")
                listing_response = requests.get(listing_url)
                listing_soup = BeautifulSoup(listing_response.content, 'html.parser')

                # Extract data from the listing page and write it to a CSV file
                title_element = listing_soup.find('h1', {'class': 'AdTitle__AdTitle-sc-1wu3o0e-0 ixxWtu'})
                if title_element is not None:
                    title = title_element.text.strip()
                else:
                    title = ""

                price_element = listing_soup.find('p', {'class': 'Price__CurrentPrice-sc-e0e8wj-0 jsgPAs'})
                if price_element is not None:
                    price = price_element.text.strip()
                else:
                    price = ""

                key_info_element = listing_soup.find('div', {'class': 'InfoPanel__KeyInfoListContainer-sc-1j25gal-4 cjGXsF'})
                if key_info_element is not None:
                    make_element = key_info_element.find('div', {'class': 'KeyInfoList__Text-sc-sxpiwc-2 dGjCWx'}, text='Make')
                    if make_element is not None:
                        make = make_element.find_next('div', {'class': 'KeyInfoList__Text-sc-sxpiwc-2 dQdfES'}).text.strip()
                    else:
                        make = ""

                    model_element = key_info_element.find('div', {'class': 'KeyInfoList__Text-sc-sxpiwc-2 dGjCWx'}, text='Model')
                    if model_element is not None:
                        model = model_element.find_next('div', {'class': 'KeyInfoList__Text-sc-sxpiwc-2 dQdfES'}).text.strip()
                    else:
                        model = ""

                    year_element = key_info_element.find('div', {'class': 'KeyInfoList__Text-sc-sxpiwc-2 dGjCWx'}, text='Year')
                    if year_element is not None:
                        year = year_element.find_next('div', {'class': 'KeyInfoList__Text-sc-sxpiwc-2 dQdfES'}).text.strip()
                    else:
                        year = ""
                    
                    mileage_element = key_info_element.find('div', {'class': 'KeyInfoList__Text-sc-sxpiwc-2 dGjCWx'}, text='Mileage')
                    if mileage_element is not None:
                        mileage = mileage_element.find_next('div', {'class': 'KeyInfoList__Text-sc-sxpiwc-2 dQdfES'}).text.strip()
                    else:
                        mileage = ""

                    transmission_element = key_info_element.find('div', {'class': 'KeyInfoList__Text-sc-sxpiwc-2 dGjCWx'}, text='Transmission')
                    if transmission_element is not None:
                        transmission = transmission_element.find_next('div', {'class': 'KeyInfoList__Text-sc-sxpiwc-2 dQdfES'}).text.strip()
                    else:
                        transmission = ""
                    fuel_type_element = key_info_element.find('div', {'class': 'KeyInfoList__Text-sc-sxpiwc-2 dGjCWx'}, text='Fuel Type')
                    if fuel_type_element is not None:
                        fuel_type = fuel_type_element.find_next('div', {'class': 'KeyInfoList__Text-sc-sxpiwc-2 dQdfES'}).text.strip()
                    else:
                        fuel_type = ""

                    body_type_element = key_info_element.find('div', {'class': 'KeyInfoList__Text-sc-sxpiwc-2 dGjCWx'}, text='Body Type')
                    if body_type_element is not None:
                        body_type = body_type_element.find_next('div', {'class': 'KeyInfoList__Text-sc-sxpiwc-2 dQdfES'}).text.strip()
                    else:
                        body_type = ""

                    color_element = key_info_element.find('div', {'class': 'KeyInfoList__Text-sc-sxpiwc-2 dGjCWx'}, text='Colour')
                    if color_element is not None:
                        color = color_element.find_next('div', {'class': 'KeyInfoList__Text-sc-sxpiwc-2 dQdfES'}).text.strip()
                    else:
                        color = ""

                    engine_size_element = key_info_element.find('div', {'class': 'KeyInfoList__Text-sc-sxpiwc-2 dGjCWx'}, text='Engine Size (Litres)')
                    if engine_size_element is not None:
                        engine_size = engine_size_element.find_next('div', {'class': 'KeyInfoList__Text-sc-sxpiwc-2 dQdfES'}).text.strip()
                    else:
                        engine_size = ""

                    nct_element = key_info_element.find('div', {'class': 'KeyInfoList__Text-sc-sxpiwc-2 dGjCWx'}, text='NCT Expiry')
                    if nct_element is not None:
                        nct_expiry = nct_element.find_next('div', {'class': 'KeyInfoList__Text-sc-sxpiwc-2 dQdfES'}).text.strip()
                    else:
                        nct_expiry = ""

                    road_tax_element = key_info_element.find('div', {'class': 'KeyInfoList__Text-sc-sxpiwc-2 dGjCWx'}, text='Road Tax')
                    if road_tax_element is not None:
                        road_tax = road_tax_element.find_next('div', {'class': 'KeyInfoList__Text-sc-sxpiwc-2 dQdfES'}).text.strip()
                    else:
                        road_tax = ""
                else:
                    make = ""
                    model = ""
                    year = ""
                    mileage = ""
                    transmission = ""
                    fuel_type = ""
                    body_type = ""
                    color = ""
                    engine_size = ""
                    nct_expiry = ""
                    road_tax = ""

                # Print the data
                print(f"Price: {price},Year: {year}, Make: {make}, Model: {model}, Mileage: {mileage}, Transmission: {transmission},Fuel Type: {fuel_type}, Body Type: {body_type}, Color: {color}, Engine Size: {engine_size}, NCT Expiry: {nct_expiry}, Road Tax: {road_tax}")
                writer.writerow({
                    "Price": price,
                    "Year": year,
                    "Make": make,
                    "Model": model,
                    "Mileage": mileage,
                    "Transmission": transmission,
                    "Fuel Type": fuel_type,
                    "Body Type": body_type,
                    "Color": color,
                    "Engine Size": engine_size,
                    "NCT Expiry": nct_expiry,
                    "Road Tax": road_tax
                })


                
```
